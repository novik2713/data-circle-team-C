```python
print("Here are the Null Features...")
for column_name in survey_raw_df.columns:
  null_count = survey_raw_df[column_name].isnull().sum()
  if null_count>0:
  #if survey_raw_df[column_name].isnull().any():     print(f"Column '{column_name}'")
    print(f"Column '{column_name}' has {null_count} null values(s)")
```

```python
#Let's represent this in a grapch
null_counts = survey_raw_df.isnull().sum()
sorted_columns= null_counts.sort_values(ascending=False)

plt.figure(figsize=(12,20))
#null_counts.plot(kind='barh', color='skyblue')
sorted_columns.plot(kind='barh', color='skyblue')
plt.xlabel=('Number of Null Vals')
plt.ylabel=('Columns')
plt.title('Null Column Distribution')

plt.show()
```

```python
# Calculate null counts for each column
null_counts = survey_raw_df.isnull().sum()

# Sort columns based on null counts
sorted_columns = null_counts.sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(20, 20))
bar_plot2 = sorted_columns.plot(kind='bar', color='skyblue')
#plt.title('Count of Null Values in Each Column')
#plt.xlabel('Columns')
#plt.ylabel('Count of Null Values')
plt.xticks(rotation=45, ha='right')

# Annotate each bar with its value
for i, val in enumerate(sorted_columns):
    bar_plot2.text(i, val, str(val), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
